# Stacks and Queues (continued)

We know from last class that a Stack is an ordered class of elements where itesmare added and removed from the top.

*Queues* are ordered collections of elements where items are added to the back and removed(and returned) from the front. 

```python
# From scartch implementation of Stack

class Stack:
    class __Node:
        def __init__(self,datum):
            self.datum = datum
            self.below = None

    def __init__(self):
        self.top = None
        self.count = 0

    def push(self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.top:
            self.top = new_node
        else:
            new_node.below = self.top
            self.top = new_node

    def pop(self):
        if self.top:
            self.count -= 1
            backup = self.top.datum
            self.top = self.top.below
            return backup
        raise IndexError("Stack is empty")

    # Nice to have methods
    def peek(self):
        # This can be solved in 0(1)
        if self.top:
            return self.top.datum
        raise IndexError("Stack is empty")

    def size(self):
        # This can be solved in 0(n)
        # Extra challenge: Tweak this Stack class succh that this has a worst case time complexity of 0(1)
        # current = self.top
        # count = 0
        # while current:
            # current = current.below
            # count += 1
        # return count
        return self.size

    def is_empty(self):
        # This can be solved in 0(1)
        return self.top is None
```
